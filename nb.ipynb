```python
df_country = pd.read_excel('.\zomato-restaurants-data\Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left') #merge two dataframes
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')
```

Major bussiness is in India, USA and UK

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
rating_color=final_df['Rating color'].value_counts().index
rating_color
```

```python
rating_count=final_df['Rating color'].value_counts().values
```

```python
plt.pie(rating_count, labels=rating_color, autopct='%1.2f%%')
```

```python
ratings
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings,palette=['white', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

```python
#Count plot
sns.countplot(x='Rating color',  data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```
